```python
#pip install netCDF4
from netCDF4 import Dataset
#pip install matplotlib
import matplotlib
#The glob module finds all the pathnames matching a specified pattern
import glob

import sys
import os.path
```

```python
#Find all zdf point cloud images in a directory
fileList=glob.glob('C:/Users/.../*.zdf')         

#Loop through all found 3D point cloud files
for FilenameZDF in fileList:
    #read the file
    data = Dataset(FilenameZDF, "r")
    # Extracting the point cloud
    pc = data["data"]["pointcloud"][:, :, :]
    #print(type(pc))
    # Extracting the RGB image
    image = data["data"]["rgba_image"][:, :, :3]
    #print(type(image))
    fileName, fileExtension=os.path.splitext(img)
    #Save the cropped image
    matplotlib.image.imsave(fileName+'_ZDFtoJPG.jpg', image)
    
    
```
